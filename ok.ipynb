```python
import pandas as pd
iris=pd.read_csv("c:\\dataset\\iris2.csv")
import matplotlib.pyplot as plt
pt.scatter(iris.Setosa_SL, iris.Setosa_SW,marker='o', color='red', label='setosa')
pt.scatter(iris.Versicolor_SL, iris.Versicolor_SW,marker='H', color='green', label='versicolor')
pt.scatter(iris.Virginica_SL, iris.Virginica_SW,marker='h', color='blue', label='virginica')
```

```python
pt.scatter (iris.Setosa_SL, iris.Setosa_SW,marker='o', color='red', label='setosa')
pt.scatter (iris.Versicolor_SL, iris.Versicolor_SW,marker='o' ,color='green', label='versicolor')
pt.scatter (iris.Virginica_SL, iris.Virginica_SW,marker='o', color='blue', label='Virginia')
pt.legend (loc='upper right')
pt.title('Iris Data')
pt.xlabel(' Sepal Length (cm)')
pt.ylabel('Sepal Width (cm) ')
```

```python
pt.scatter (iris.Setosa_SL, iris.Setosa_SW,marker='o', color='red', label='setosa')
pt.scatter (iris.Versicolor_SL, iris.Versicolor_SW,marker='o' ,color='green', label='versicolor')
pt.scatter (iris.Virginica_SL, iris.Virginica_SW,marker='o', color='blue', label='Virginia')
pt.legend (loc='upper right')
pt.title('Iris Data')
pt.xlabel(' Sepal Length (cm)')
pt.ylabel('Sepal Width (cm) ')

pt.annotate('setosa', xy=(5.0, 3.5))
pt.annotate ('virginica', xy=(7.25, 3.5))
pt.annotate ('versicolor', xy=(5.0, 2.0))
```
