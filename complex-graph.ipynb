```python
import numpy as nu
import matplotlib.pyplot as plt

import seaborn as sbn
heightwei=sbn.load_dataset('htwt')
```

```python
plt.subplot(1,2,1)
sbn.boxplot(x='sex',y='height', data=heightwei)
plt.ylabel('height(cm)')

plt.subplot(1,2,2)
sbn.violinplot(x='sex',y='height', data=heightwei)
plt.ylabel('height(cm)')
plt.tight_layout()
```

```python
sbn.violinplot(x='sex',y='height',data=heightwei,inner=None,color='lightgray')
sbn.stripplot(x='sex',y='height',data=heightwei,size=9,jitter=True)
```

```python
sbn.jointplot(x='height',y='weight',data=heightwei,kind='kde')
```

```python
sbn.jointplot(x='height',y='weight',data=heightwei)
```

```python
sbn.pairplot(heightwei)
```

```python
sbn.pairplot(heightwei,hue='sex')
```

```python
cov_hw=nu.corrcoef(heightwei['height'],heightwei['weight'])
print(cov_hw)
```

```python
sbn.heatmap(cov_hw)
```

```python
hkex=pd.read_csv("c:\\dataset\\0388HK.csv",index_col=0)
cp=hkex['Close']
d10=cp.rolling(window=10).mean()
d20=cp.rolling(window=20).mean()
d50=cp.rolling(window=50).mean()
d100=cp.rolling(window=100).mean()
d150=cp.rolling(window=150).mean()
d200=cp.rolling(window=200).mean()
d250=cp.rolling(window=250).mean()
mvv={'close':cp,'d10':d10,'d20':d20,'d50':d50,'d100':d100,'d150':d150,'d200':d200,'d250':d250}
exmv=pd.DataFrame(mvv)
exmv=exmv['2018-01':'2019-02']
exmv.tail(10)
```

```python
plt.plot(exmv['close'],color='blue',label='1')
plt.plot(exmv['d10'],color='red',label='1')
plt.plot(exmv['d20'],color='yellow',label='1')
plt.plot(exmv['d50'],color='green',label='1')
plt.plot(exmv['d100'],color='purple',label='1')
plt.plot(exmv['d150'],color='gray',label='1')
plt.legend(loc='upper center')
plt.xticks(rotation=60)
```
