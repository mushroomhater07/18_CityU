```python
import seaborn as sbn
import matplotlib.pyplot as plt
import pandas as pd

heightwei=sbn.load_dataset('htwt')
sbn.lmplot(x='weight',y='height', data=heightwei , col='sex')
```

```python
sbn.residplot(x='weight',y='height',data=heightwei,color='indianred')
```

```python
sbn.stripplot(y='height',data=heightwei)
plt.ylabel('height(cm)')
```

```python
sbn.stripplot(y='height',data=heightwei,jitter=False)
plt.ylabel('height(cm)')
```

```python
sbn.stripplot(x='sex',y='height',data=heightwei,jitter=False)
plt.ylabel('height(cm)')
```

```python
sbn.swarmplot(x='sex',y='height',data=heightwei)
plt.ylabel('height(cm)')
```

```python
sbn.swarmplot(y='sex',x='height',data=heightwei,hue='sex',orient='h')
plt.ylabel('agegroup')
plt.xlabel('height(cm)')
plt.tight_layout()
```

```python
sbn.swarmplot(y='sex',x='height',data=heightwei,hue='sex')
plt.ylabel('agegroup')
plt.xlabel('height(cm)')
plt.tight_layout()
```
